```r
library(TSA)
```

```r
data(tempdub)
```

```r
length(tempdub)
```

```r
lm <- lm(tempdub~sin(2*pi*time(tempdub)) + cos(2*pi*time(tempdub)))
```

```r
summary(lm)
```

```r
plot(ts(fitted(lm), freq=12, start=c(1964,1)), col="red")
points(tempdub)
```

```r
NYBirths <- read.table("nybirths.dat.txt")
NYBirths <- ts(NYBirths, frequency = 12, start = c(1946, 1))
names(NYBirths) <- "MonthlyBirths"
```

```r
plot.ts(NYBirths)
```

```r
plot(decompose(NYBirths))
```

```r
lm2 <- lm(NYBirths ~ sin(2*pi*time(NYBirths)) + cos(2*pi*time(NYBirths)))
summary(lm2)
```

```r
plot(ts(fitted(lm2), freq=12, start=c(1946, 1)), col="red")
points(NYBirths)
```
