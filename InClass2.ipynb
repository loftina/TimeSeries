```r
observations = 1000
randomWalk <- rep(0,observations)
randomWalk[1] <- 2
for (i in 2:observations){
   randomWalk[i] <- randomWalk[i-1] + rnorm(n = 1) 
}
```

```r
randomWalk <- ts(randomWalk)
plot.ts(randomWalk, ylab="Random", main="Random Walk with Standard Normal Distribution", col = 'blue')
abline(h=2, col = 'red')
```

```r
noise <- rnorm(101, mean = 0, sd = 1)
movingAverage <- rep(0, 100)

for (i in 2:101) {
    movingAverage[i-1] = (noise[i] + noise[i-1])/2
}
```

```r
movingAverage <- ts(movingAverage)
plot(movingAverage)
```

```r
count = 0
for (i in 1:100){
    if (movingAverage[i] > -1.96*sqrt(1/2) && movingAverage[i] < 1.96*sqrt(1/2)) {
        count = count + 1
    }
}
count
```
