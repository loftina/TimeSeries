<p style="font-size:20px; text-decoration:underline;">
Importing Data:
</p>

```r
library(TSA)

AAPL <- read.csv("AAPL.csv", header = T)
AAPL.AdjClose <- ts(AAPL$Adj.Close)
```

<p style="font-size:20px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This dataset is 5 years worth of adjusted closing prices for Apple stock from 9/12/2013 to 9/11/2018.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The dataset consists of 1259 floating point values.<br><br>
<div style="font-size:20px; text-decoration:underline;">Summary statistics:</div>
</p>

```r
cat("Mean: ", mean(AAPL.AdjClose), "\n")
cat("Standard Deviation: ", sd(AAPL.AdjClose), "\n")
cat("Median: ", median(AAPL.AdjClose), "\n")
cat("Variance: ", var(AAPL.AdjClose), "\n")
```

<p style="font-size:20px; text-decoration:underline;">
Plotting Adjusted Closing Prices:
</p>

```r
plot(AAPL.AdjClose,
    ylab="Adjusted Closing Pice ($)",
    xlab="Time",
    main="Apple Adjusted Closing Price 9/12/2013 - 9/11/2018")
```

<p style="font-size:20px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;These data show an upward trend in the adjusted closing price of Apple stock, meaning the mean tends to increase with time.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Variance appears to remain reasonably constant over time.
</p>

<p style="font-size:20px; text-decoration:underline;">
Checking For Outliers:
</p>

```r
hist(AAPL.AdjClose)
boxplot(AAPL.AdjClose)
```

<p style="font-size:20px;">
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The boxplot shows that 4 observations are outliers. These observations are likely due to the large, recent spike in the stock price, and may not be indicative of the true, overall trend of the stock.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The histogram is positively skewed.
</p>

<p style="font-size:20px; text-decoration:underline;">
Checking for Autocorrelation With Lag of 1 Timestep:
</p>

```r
plot(AAPL.AdjClose,
     zlag(AAPL.AdjClose),
    ylab="Adjusted Close at t",
    xlab="Adjusted Close at t-1",
    main="Correlation between successive days")
```

```r
cor(AAPL.AdjClose[2:length(AAPL.AdjClose)], na.omit(zlag(AAPL.AdjClose)))
```

<p style="font-size:20px;">
The plot and corellation of .9989 suggest that successive data points are highly correlated.
</p>

<p style="font-size:20px; text-decoration:underline;">
Linear Model for Stock Price:
</p>

```r
lm <- lm(AAPL.AdjClose ~ time(AAPL.AdjClose))

plot(AAPL.AdjClose,
    ylab="Adjusted Closing Pice ($)",
    xlab="Time",
    main="Apple Adjusted Closing Price 9/12/2013 - 9/11/2018")
abline(lm)

summary(lm)
qqnorm(resid(lm))
qqline(resid(lm))

plot(resid(lm), col='blue', ylab="Residuals")
abline(h=0, col='red')
```

<p style="font-size:20px;">
The QQ Plot suggests that the error term is not normally distributed due to the skew on the ends of the plot.
Furthermore, the residual plot has a very apparent pattern rather than being closely and evenly distributed about the mean. These findings suggest that a linear model is a very poor fit for these data.<br>
On the other hand, the f statistic is very statistically significant with a p-value of $<2.2*10^{-16}$
</p>

<p style="font-size:20px; text-decoration:underline;">
Linear Model for Lag of 1:
</p>

```r
lm <- lm(AAPL.AdjClose[2:length(AAPL.AdjClose)] ~ na.omit(zlag(AAPL.AdjClose)))

plot(AAPL.AdjClose,
     zlag(AAPL.AdjClose),
    ylab="Adjusted Close at t",
    xlab="Adjusted Close at t-1",
    main="Correlation between successive days")
abline(lm)

summary(lm)

qqnorm(resid(lm))
qqline(resid(lm))

plot(resid(lm), col="blue", ylab="Residuals")
abline(h=0, col="red")
```

<p style="font-size:20px;">
There are a few flaws with this model. The intercept value is not statistically significant and the qq plot shows a skew on both ends of the plot. Also, the residual plot shows an uneven distribution about the x axis.<br>
Despite these flaws, the F Statistic is very statistically significant with its p-value of $<2.2*10^{-16}$
</p>
