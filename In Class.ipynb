```r
install.packages("TSA")
library(TSA)
```

```r
data(larain)
class(larain)
plot(larain, ylab="Inches", xlab="Year", main="Annual rainfall data in LA", type="o")
plot(y=larain, x=zlag(larain), ylab="Inches", xlab="Previous dataset", main="Correlation")
```

```r
cor(y=larain[2:length(larain)], x=na.omit(zlag(larain)))
```
