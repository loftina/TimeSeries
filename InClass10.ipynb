```r
par(mfrow=c(2,1))
library(TSA)
```

```r
alpha1 = .6
alpha2 = .4
theta1 = .6
theta2 = .4
```

<center>ACF for MA(1) and AR(1)</center>

```r
noise <- rnorm(100)
MA1 <- c()
MA1[1] <- noise[1]

for (t in 2:100){
    MA1[t] = noise[t] + alpha1*noise[t-1]
}
```

```r
AR1 <- c()

AR1[1] <- noise[1]

for (t in 2:100){
    AR1[t] <- theta1*AR1[t-1] + noise[t]
}
```

```r
par(mfrow=c(2,1))
plot.ts(MA1)
plot.ts(AR1)
```

```r
par(mfrow=c(2,1))
acf(MA1)
acf(AR1)
```

```r
noise <- rnorm(100)
MA2 <- c()
MA2[1] <- noise[1]
MA2[2] <- noise[2]

for (t in 3:100){
    MA2[t] = noise[t] + alpha1*noise[t-1] + alpha2*noise[t-2]
}
```

```r
AR2 <- c()

AR2[1] <- noise[1]
AR2[2] <- noise[2]

for (t in 3:100){
    AR2[t] <- theta1*AR2[t-1] + theta2*AR2[t-2] + noise[t]
}
```

```r
par(mfrow=c(2,1))
plot.ts(MA2)
plot.ts(AR2)
```

```r
acf(MA2)
acf(AR2)
```

```r
pacf(AR1)
pacf(AR2)
```

<center>Theta of 1 means non stationary. ACF function shows linear decay instead of exponential</center>

```r
AR1s <- c()

AR1s[1] <- noise[1]

for (t in 2:100){
    AR1s[t] <- AR1s[t-1] + noise[t]
}
```

```r
acf(AR1s)
```
