```r
alpha1 = .8
theta1 = .6
```

```r
Y_t <- c()
noise <- rnorm(100)

Y_t[1] <- noise[1]

for (t in 2:100){
    Y_t[t] <- alpha1*Y_t[t-1]+theta1*noise[t-1]+noise[t]
}
```

```r
plot.ts(Y_t)
```

```r
par(mfrow=c(2,1))
acf(Y_t, lag.max = 50)
pacf(Y_t, lag.max = 50)
```
