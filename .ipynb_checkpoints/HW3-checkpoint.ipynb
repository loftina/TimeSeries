1. Suppose cov(Xt,Xtâˆ’k) = gammak is free of t but that E(Xt) = 3t.<br>

(a) Is Xt stationary?<br>
No, the expected value must be independent of t<br>
  
(b) Let Yt=7-3t+Xt. Is Yt stationary?<br>
E[Yt] = E[7-3t+Xt] = 7 - 3t + 3t = 7<br>

var(Yt) = var(7-3t+Xt) = var(Xt) = cov(Xt, Xt) = gamma0<br>

cov(Yt, Yt-k) = cov(7-3t+Xt, 7-3t+Xt-k) = cov(Xt, Xt-k) = gammak<br>

Yes, Yt is stationary because mean, variance, and autocovariance are all independent of t<br>

(c) Simulate 100 observations for both Xt and Yt and plot as a time series. Check if your answer for parts a and b are consistent with your simulation.<br>

```r
set.seed(2018)
time <- 1:100
noise <- rnorm(100, mean=0, sd=10)
X_t <- c()

for (t in time) {
  X_t[t] <- 3*t + noise[t]
}

plot.ts(X_t)


Y_t <- c()
for (t in time){
  Y_t[t] <- 7 - 3*t + X_t[t]
}

plot.ts(Y_t)
```

Simulation is consistent with analytical solution.<br>

2. Let Xt be a stationary time series, and define Yt = Xt when t is odd and Xt + 3 when t is even<br>

(a) Show that cov(Yt,Yt-k) is free of t for all lags k.<br>
odd t and k<br>
cov(Xt, Xt-k + 3) = cov(Xt, Xt-k) = gammak<br>
odd k even t or odd t even k<br>
cov(Xt, Xt-k) = gammak<br>
even t and k<br>
cov(Xt + 3, Xt-k + 3) = cov(Xt, Xt-k) = gammak<br>

(b) Is Yt stationary?<br>
t is odd<br>
E[Yt] = E[Xt] = constant<br>
t is odd<br>
E[Yt] = E[Xt+3] = 3 + constant<br>

var(Yt) = var(Xt + 3) = var(Xt) = constant<br>

Yes, Yt is stationary because mean, variance, and autocovariance are all independent of t<br>

(c) Plot 100 simulated observations for the random variable Yt and comment on your finding.<br>

```r
X_t <- c()
for (t in time) {
  X_t[t] <- rnorm(1)
}

Y_t <- c()
for (t in time) {
  if (t %% 2 == 0) {
    Y_t[t] <- X_t[t]
  }
  else {
    Y_t[t] <- X_t[t] + 3
  }
}

plot.ts(X_t)
plot.ts(Y_t)
```

Simulation shows Yt is indeed stationary using Xt~N(0,1)<br>

3. Random Cosine Wave Let Yt = cos(2pi(t/12 + phi)) where phi is selected from a uniform distribution on the interval 0 to 1.

(a) Find the expected value and variance of Yt<br>

Let I(x) means integral of x from 0 to 1<br>
E[Yt] = I(Yt)dphi = I(cos(2pi(t/12 + phi))) dphi  = 1/2pi sin[2pi(t/12)+phi] evaluated from 0 to 1<br>
= 1/2pi [sin(2pi(t/12+1)) - sin(2pi(t/12))] = 1/2pi[0] = 0<br>

Var[Yt] = E[Yt^2] - (E[Yt])^2 = E[Yt^2] = I(cos^2(2pi(t/12+phi)))dphi<br>
= 1/2 I(cos(4pi(t/12+Ï†))+1)dphi = 1/2[1/4pi sin(4pi(t/12+phi)) + 1] evaluated from 0 to 1<br>
= 1/8pi[sin(4pi(t/12+1)) - sin(4pi(t/12))] + 1/2 = 1/2<br>

(b) plot 100 simulated observations and check if there is any trend.<br>

```r
Y_t <- c()
noise <- runif(100, min = 0, max = 1)
for (t in time) {
  Y_t[t] <- cos(2*pi*(t/12 + noise[t]))
}

plot.ts(Y_t)
```
