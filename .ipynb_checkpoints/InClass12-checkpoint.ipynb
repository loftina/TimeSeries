```r
library(quantmod)
```

```r
getSymbols("AAPL", from="2016-01-01", to="2018-01-01")
head(AAPL)
```

```r
class(AAPL)
```

```r
plot(AAPL$AAPL.Close)
```

```r
barChart(AAPL, theme="white", bar.type="hlc")
```

```r
candleChart(AAPL, multi.col = T, theme = "white")
```

```r
chartSeries(AAPL, theme = "white", TA=c(addSMA(20, col = "red"), addSMA(50, col="orange"), addSMA(100, col="yellow"), addVo()), type = "l")
```

```r
Closing <- AAPL$AAPL.Close
SMA20 = SMA(Closing, n=20)
SMA50 = SMA(Closing, n=50)
SMA100 = SMA(Closing, n=100)
plot(Closing)
lines(SMA20, col="red")
lines(SMA50, col="orange")
lines(SMA100, col="yellow")
```
