<h1>Stationarity</h1>
A stochastic time series X_t is starionary if for any positive number k the distribution of X_t1, X_t2,..., X_tn is the same as X_t1-k, X_t2-k,..., X_tn-k

```r
library(TSA)
```

```r
X_t <-(rnorm(100))
X_t <- cumsum(X_t)
X_t <- ts(X_t)
cor(X_t[2:length(X_t)], na.omit(zlag(X_t)))
```

```r
cor(X_t[2:25], na.omit(zlag(X_t[1:25])))
```

```r
X_t <- rnorm(100)
MA <- rep(0, 100)
MA[1] <- X_t[1]
for (t in 2:100) {
    MA[t] <- .5*(X_t[t] + X_t[t-1])
}
MA <- ts(MA)
```

```r
cor(MA[2:length(MA)], na.omit(zlag(MA)))
```

```r
y <- MA[2:length(MA)]
x <- na.omit(zlag(MA))
```

```r
data11 <- data.frame(y=y, x=x)
ts.plot(data11, col=rainbow(2))
```

```r
y2 <- MA[3:length(MA)]
x2 <- na.omit(zlag(MA, 2))
data22 <- data.frame(y=y2, x=x2)

ts.plot(data22,col=rainbow(2))
```
