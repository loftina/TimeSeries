```r
MortalityRate <- read.table("MortalityRate.txt", header = T)
head(MortalityRate)
```

```r
LatSquared = MortalityRate$Lat^2
quadraticModel <- lm(MortalityRate$Mort ~ MortalityRate$Lat + LatSquared)
summary(quadraticModel)
```

```r
plot(x = MortalityRate$Lat, y = MortalityRate$Mort)
abline(quadraticModel)
```

Let x_t~N(0, 4)<br>
and Y_t = 5+2t+x_t<br>
t=1,2,...,100

```r
Y_t <- rep(0, 100)
time <- 1:100
for (t in time){
    Y_t[t] <- 5 + 2*t + rnorm(n = 1, mean = 0, sd = 1)
}
Y_t = ts(Y_t)
plot(Y_t)
```

```r
model0 <- lm(Y_t ~ time)
summary(model0)
cor(Y_t, time)
```

```r
plot(Y_t)
abline(model0, col="red")
```
